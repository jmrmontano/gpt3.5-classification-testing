## Dataset 2

```python
label_counts = df['prognosis'].value_counts()
print(label_counts)
```

```python
df1 = df[df['prognosis'] == 'Hepatitis C']
df2 = df[df['prognosis'] == 'Hepatitis B']
df3 = df[df['prognosis'] == 'Hepatitis D']
```

```python
df_total= pd.concat([df1, df2, df3], axis=0)
```

```python
label_counts = df_total['prognosis'].value_counts()
print(label_counts)
```

## New Dataset

```python

df_total.to_csv('df_total.csv', index=False)
```
