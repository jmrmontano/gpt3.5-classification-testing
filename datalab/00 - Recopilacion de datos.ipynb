## Libreria

```python
import pandas as pd
```

## Configuracion

```python
SAN_IGNACIO_DATASET_URL = 'https://drive.google.com/uc?id=1fQ_SMYW8nE_lts1QMus9HKZ2iSifEoYC&confirm=t' # archivo datos_san_ignacio.csv - originales
```

## Carga de datos

```python
san_ignacio_dataframe = pd.read_csv(SAN_IGNACIO_DATASET_URL)
```

```python
san_ignacio_dataframe.head()
```

```python
san_ignacio_dataframe.shape
```

## Organizacion de columnas

```python
san_ignacio_dataframe.rename(columns={'idCliente':'IdCliente', 'fechaIngresoAtencion':'FechaIngresoAtencion', 'sexo':'Sexo', 'fecNacimiento':'FechaNacimiento', 
                            'desPlanYConcepto':'AnalisisyPlandeManejo', 'codCie':'CodDx', 'NomDiagnostico':'NombreDx'}, inplace=True)

san_ignacio_dataframe= san_ignacio_dataframe[['IdCliente','FechaIngresoAtencion','Sexo','CodDx','NombreDx','FechaNacimiento','AnalisisyPlandeManejo']]
```

```python
san_ignacio_dataframe.head()
```

## Homogenizacion de datos

```python
san_ignacio_dataframe.loc[san_ignacio_dataframe.Sexo == 'Femenino', 'Sexo'] = 'F'
san_ignacio_dataframe.loc[san_ignacio_dataframe.Sexo == 'f', 'Sexo'] = 'F'
san_ignacio_dataframe.loc[san_ignacio_dataframe.Sexo == 'm', 'Sexo'] = 'M'
```

## Identifiacion valores nulos

```python
san_ignacio_dataframe.isna().sum()
```

```python
san_ignacio_dataframe.dropna(inplace=True)
```

```python
san_ignacio_dataframe.isna().sum()
```

```python
san_ignacio_dataframe.shape
```

## Creación de etiquetas en columna "label"

```python
# Identificación de observaciones con APNEA. se genera la columna "label". 1 = Si diagnostico y 0 = No Diagnostico
san_ignacio_dataframe['label'] = ["1" if s == "APNEA DEL SUEÑO" else "0" for s in san_ignacio_dataframe['NombreDx']] 
```

## Seleccion de datos estadisticamente significativos

Muestreo estratificado

```python
san_ignacio_dataframe.head(5)
```

```python
conteo_por_label = san_ignacio_dataframe['label'].value_counts()
print(conteo_por_label)
```

```python
# seleccionar una muestra aleatoria estratificada
muestras_estratificadas = san_ignacio_dataframe.groupby('label').apply(lambda x: x.sample(n=20))
```

```python
numero_de_estratos = muestras_estratificadas['label'].nunique()
print(numero_de_estratos)
```

```python
conteo_valores = muestras_estratificadas['label'].value_counts()
print(conteo_valores)
```
